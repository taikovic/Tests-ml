# Regression polynomial

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

# Etape 1: generer les datas: 
m=100
x=6 * np.random.rand(m,1) - 3
y= 2 * x * x + x + np.random.randn(m,1)

plt.scatter(x,y)
```

# Application Scikit Learn

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
#Ajoutons les termes du second degree: 2 

poly_features = PolynomialFeatures(degree=2, include_bias=False)
#PolynomialFeatures(degree=n) transform la matrice de n variables en (n+d)!/n!d! variables

X_poly = poly_features.fit_transform(x)
X_poly[0] # cette variable contient la premiere variable x[0] et son carre

#application de la regression poly:

lin_regPoly=LinearRegression()
lin_regPoly.fit(X_poly,y)
lin_regPoly.intercept_, lin_regPoly.coef_
```

## Plot 

```python
plt.plot(X_poly,y,"b.")
plt.plot(X_poly,lin_regPoly.predict(X_poly),color='R')
plt.show()
```
